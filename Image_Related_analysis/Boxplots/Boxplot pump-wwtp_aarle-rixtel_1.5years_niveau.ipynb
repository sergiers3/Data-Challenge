```python
import pandas as pd
```

```python
df1 = pd.read_csv("niveau1.csv",sep=";", header=0)
```

```python
d = {"Tag_Name": ["GBS_AR.RG4110/Instellingen/LT610/Niveaumeting.Q",
"GBS_AR.RG4110/Proces/LS411/Niveaumeting.Q", "GBS_AR.RG4140/Instellingen/LT610/Niveaumeting.Q", "GBS_AR.RG4150/Instellingen/LT610/Niveaumeting.Q", "GBS_AR.RG4160/Instellingen/LT610/Niveaumeting.Q", "GBS_AR.RG4170/Instellingen/LT610/Niveaumeting.Q","GBS_AR.RG4190/Instellingen/LT610/Niveaumeting.Q", "GBS_AR.RG4190/Proces/LT620/Niveaumeting.Q"], "Pumps": ["P4","P8", "P5", "P6", "P2", "P1", "P7", "P3"]}
dftable = pd.DataFrame(data=d)
result = df1.merge(dftable, on="Tag_Name")
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
fig, ax = plt.subplots()
fig.set_size_inches(10, 13)

sns.boxplot(ax=ax, x='Pumps', y='Value', data=result,order=["P1","P2","P3","P4","P5","P6","P7","P8"])
```
