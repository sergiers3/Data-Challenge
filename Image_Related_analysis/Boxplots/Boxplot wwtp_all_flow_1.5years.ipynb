```python
import glob
import pandas as pd
path =r'C:\Users\Dagoberto\data_wwtp_all_flow_1.5years-20181015T155110Z-001' # use your path
allFiles = glob.glob(path + "/*.csv")
frame = pd.DataFrame()
list_ = []
for file_ in allFiles:
    df = pd.read_csv(file_,sep=";", header=0)
    list_.append(df)
frame = pd.concat(list_)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots()
fig.set_size_inches(10, 13)
sns.boxplot(ax=ax, x='Area_Name', y='Value', data=frame,order=["Aarle", "Oijen","Dinther","Land van Cuijk","Vinkel"])
```
